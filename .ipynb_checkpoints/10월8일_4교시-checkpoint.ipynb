```python
import time
from selenium import webdriver #가상 크롬 드라이버를 돌린다
from selenium.webdriver.chrome.options import Options #가상 크롭 드라이버의 옵션값 설정
from selenium.webdriver.chrome.service import Service #가상 크롭 드라이버를 자동 설치
from webdriver_manager.chrome import ChromeDriverManager #자동 설치할 때 쓰는 매니지먼트
from selenium.webdriver.common.keys import Keys #키보드에서 어떤 키를 입력했는지 알아냄 (엔터, 탭, esc.. 등등 어케 표현할거임?)
from selenium.webdriver.common.by import By #태그로 가져와 css로 가져와..?
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

options=Options()
options.add_experimental_option('detach',True) #화면이 꺼지지 않고 유지
options.add_argument('--start-maximized') # 화면을 최대화
service=Service(ChromeDriverManager().install()) #크롬드라이버 자동 설치

driver=webdriver.Chrome(service=service, options=options)


def naver_blog(search,pages=1):
    url=f'https://section.blog.naver.com/Search/Post.naver?pageNo={pages}&rangeType=ALL&orderBy=sim&keyword={search}'
    driver.get(url)
    time.sleep(2) #2초 멈춤
    html=driver.page_source #현재 열려있는 창에 모든 소스를 html에 담기
    #print(html)
    soup = bs(html,'html')
    return soup
    
search='가을'
pages=3
res=[]

for i in range(1,pages+1):
    res.append(naver_blog(search,i))
    
driver.quit() #닫기버튼- 메모리에서 제거 , 맨 마지막에 하는 이유
```

```python
res_df=[]
for i in res:
    title_tags = i.select('a.desc_inner')
    for j in title_tags:
        tmp={}
        tmp['제목']=j.text.strip()
        tmp['링크']=j['href']
        res_df.append(tmp)
df=pd.DataFrame(res_df)
df
```
