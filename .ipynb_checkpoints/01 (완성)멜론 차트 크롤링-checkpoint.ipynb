<참고> User Agent: 
- 인터넷 상에서 소프트웨어가 자신을 다른 소프트웨어나 웹 서버에 소개하는 방법
- 주로 웹 브라우저나 웹 크롤러에서 사용되며, 네트워크 요청을 할 때 이를 식별하는 데 쓰이는 문자열을 포함 함
    - 브라우저의 종류와 버전: 예를 들어, Chrome 96, Firefox 34 
    - 운영 체제: 사용자의 운영 체제와 버전 정보, 예를 들어 Windows 10, macOS, Android 등입니다.
    - 디바이스 정보: 특히 모바일 디바이스에서 사용되며, 모델명이나 제조사를 나타낼 수 있습니다.

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url='https://www.melon.com/chart/day/index.htm'
html=requests.get(url)
html
```

```python
url='https://www.melon.com/chart/day/index.htm'
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'
hdr = {'User-Agent': user_agent}
html=requests.get(url,headers=hdr)
html
```

```python
soup=bs(html.text,'html.parser')
table=soup.find('table')
df=pd.read_html(str(table))[0]
df
```

## 크롬 드라이버 설치하지 않고 사용하기

* 참고 : 크롬 드라이버 설치해서 사용하는 방법  => 문제점: 크롬버전이 업그레이드할 때마다 다시 설치해야함
  https://countrymouse.tistory.com/entry/ChromeDriver

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import pandas as pd

options=Options()
options.add_experimental_option('detach',True) #화면이 꺼지지 않고 유지
options.add_argument('--start-maximized') # 화면을 최대화
service=Service(ChromeDriverManager().install()) #크롬드라이버 자동 설치

driver=webdriver.Chrome(service=service, options=options)


url='https://www.melon.com/chart/day/index.htm'
driver.get(url)
time.sleep(2)
html=driver.page_source
driver.close()   #현재 탭 닫기
driver.quit()    #브라우저 닫기
```

```python
soup=bs(html,'html.parser')
```

```python
table=soup.find('table')
df=pd.read_html(str(table))
df
```
